```python
products = ['Macbook Pro Laptop', 'LG Washing Machine', 'USB-C Charging Cable',
            '27in FHD Monitor', 'AA Batteries (4-pack)',
            'Bose SoundSport Headphones', 'AAA Batteries (4-pack)',
            'ThinkPad Laptop', 'Lightning Charging Cable', 'Google Phone',
            'Wired Headphones', 'Apple Airpods Headphones', 'Vareebadd Phone',
            'iPhone', '20in Monitor', '34in Ultrawide Monitor',
            'Flatscreen TV', '27in 4K Gaming Monitor', 'LG Dryer']

product_data = []

for product in products:
    product_df = df[df['Product'] == product]
    monthly_sales = product_df.groupby('Month')['Quantity Ordered'].sum().reset_index()  # Reset index for better DataFrame structure
    monthly_sales.rename(columns={'Quantity Ordered': f'Sales - {product}'}, inplace=True)
    product_data.append(monthly_sales)

result_df = product_data[0]
for i in range(1, len(product_data)):
    result_df = pd.merge(result_df, product_data[i], on='Month', how='outer')


result_df = result_df.fillna(0)
result_df

```
