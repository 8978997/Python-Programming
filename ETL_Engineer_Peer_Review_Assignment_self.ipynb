## Transform

Using <code>exchange_rate</code> and the `exchange_rates.csv` file find the exchange rate of USD to GBP. Write a transform function that

1. Changes the `Market Cap (US$ Billion)` column from USD to GBP
2. Rounds the Market Cap (US$ Billion)` column to 3 decimal places
3. Rename `Market Cap (US$ Billion)` to `Market Cap (GBP$ Billion)`


```python
def transform():
    exchange_rates = df.loc[df['currency'] == 'GBP', 'Rates'].values[0]
    extract2['Market Cap (GBP$ Billion)'] = extract2['Market Cap (US$ Billion)'] * exchange_rates
    extract2['Market Cap (GBP$ Billion)'] = extract2['Market Cap (GBP$ Billion)'].round(3)
    extract2.rename(columns={'Market Cap (US$ Billion)': 'Market Cap (GBP$ Billion)'}, inplace=True)
    return extract2
transform()
```
